```python
import pandas as pd

# --- Split 4th column (index 3) into branch_code and branch_name ---
split_cols = df.iloc[:, 3].str.split('-', n=1, expand=True)

df.insert(4, 'branch_code', split_cols[0])   # first part before '-'
df.insert(5, 'branch_name', split_cols[1])   # second part after '-'
df.drop(df.columns[3], axis=1, inplace=True)  # drop original column

# --- Shift everything from row 5 downward by 1 to make space ---
df.iloc[5:] = df.iloc[5:].shift(1)
df.loc[5] = None   # make sure row 5 exists and is blank

# --- For the row with index 4, split cells containing '-' into row 4 and row 5 ---
for col in df.columns:
    val = df.at[4, col]
    if isinstance(val, str) and '-' in val:
        before, after = val.split('-', 1)
        df.at[4, col] = before
        df.at[5, col] = after  # safe: row 5 is empty now

df = df.reset_index(drop=True)

df
```

```python
# Mapping from product code → new service type
def map_service_type(code):
    if code == "EP" or code == "BP":
        return "Air Red"
    elif code == "ES" or code == "BS":
        return "Air White"
    elif code == "GP":
        return "Ground"
    else:
        return None

# Insert a new column at position 6 with the new service type
df.insert(
    5,
    'service_type',
    df.iloc[:, 5].map(map_service_type)
)

df
```

```python
# From 7th row and 8th column, ensure all values are numbers, fix numbers written as text with commas and decimals, fill missing values with 0, and divide each value by 25 to get per day values

# Select the sub-DataFrame to process
numeric_block = df.iloc[6:, 7:]

# Remove commas (thousands separator), convert to float, coerce errors to NaN, and divide by 25
def clean_number_per_day(x):
    if pd.isna(x):
        return 0
    try:
        return float(str(x).replace(',', '')) / 25
    except Exception:
        return 0

numeric_block = numeric_block.applymap(clean_number_per_day)

# Assign back to the original DataFrame
df.iloc[6:, 7:] = numeric_block

df
```

```python
import pandas as pd
import json
from collections import defaultdict

csv_path = "data.csv"

# -------------------------
# --- Read CSV -----------
# -------------------------
df = pd.read_csv(
    csv_path,
    skiprows=1,               # skip empty first row
    header=[0,1,2,3,4,5],     # 6 header rows for destination columns
    index_col=[0,1,2,3,4]     # 5 index levels for origin
)

# -------------------------
# --- Name index and columns
# -------------------------
df.index.set_names(
    ["org_zone", "org_region", "org_city", "org_branch_code", "org_branch_name"],
    inplace=True
)

df.columns.set_names(
    ["type", "des_zone", "des_region", "des_city", "des_branch_code", "des_branch_name"],
    inplace=True
)

# -------------------------
# --- Build origin hierarchical mapping
# -------------------------
origin_mapping = defaultdict(lambda: defaultdict(lambda: defaultdict(dict)))

for idx in df.index:
    zone, region, city, branch_code, branch_name = idx
    origin_mapping[zone][region][city][branch_code] = branch_name

# Save origin mapping
with open("org_mappings.json", "w") as f:
    json.dump(origin_mapping, f, indent=4)

print("✅ Origin mapping saved as org_mappings.json")

# -------------------------
# --- Build destination mapping by type
# -------------------------
col_df = df.columns.to_frame(index=False)

def build_des_mapping(sub_df):
    des_mapping = defaultdict(lambda: defaultdict(lambda: defaultdict(dict)))
    for _, row in sub_df.iterrows():
        zone = row['des_zone']
        region = row['des_region']
        city = row['des_city']
        branch_code = row['des_branch_code']
        branch_name = row['des_branch_name']
        des_mapping[zone][region][city][branch_code] = branch_name
    return des_mapping

# Separate by type
volume_df = col_df[col_df['type'] == 'Volume'][['des_zone','des_region','des_city','des_branch_code','des_branch_name']].drop_duplicates()
billed_df = col_df[col_df['type'] == 'Billed Wt'][['des_zone','des_region','des_city','des_branch_code','des_branch_name']].drop_duplicates()

# Flatten for validation
flat_volume = dict(zip(volume_df['des_branch_code'], volume_df['des_branch_name']))
flat_billed = dict(zip(billed_df['des_branch_code'], billed_df['des_branch_name']))

# -------------------------
# --- Validate and save destination mapping
# -------------------------
if flat_volume == flat_billed:
    print("✅ Volume and Weight Billed branch mappings are identical")
    des_mapping = build_des_mapping(volume_df)  # only need one
    with open("des_mappings.json", "w") as f:
        json.dump(des_mapping, f, indent=4)
    print("✅ Destination mapping saved as des_mappings.json")
else:
    print("❌ Volume and Weight Billed branch mappings differ! Not saving destination mapping.")
    for k in set(flat_volume.keys()).union(flat_billed.keys()):
        v1 = flat_volume.get(k)
        v2 = flat_billed.get(k)
        if v1 != v2:
            print(f"Branch code {k}: Volume='{v1}', Billed='{v2}'")
```

```python
import pandas as pd

def summarize_by_org_branch_type_service_optimized(csv_path="data.csv"):
    # ==========================
    # Step 1: Load CSV once
    # ==========================
    df = pd.read_csv(
        csv_path,
        skiprows=1,  # skip metadata row
        header=[0, 1, 2, 3, 4, 5],  # include 6 header rows (up to des_branch_name)
        index_col=[0, 1, 2, 3, 4, 5, 6],  # origin info
        low_memory=False
    )

    # Set proper names for indexes and columns
    df.index.set_names(
        [
            "org_zone", "org_region", "org_city",
            "org_branch_code", "org_branch_name",
            "service_type", "org_product"
        ],
        inplace=True
    )
    df.columns.set_names(
        ["type", "des_zone", "des_region", "des_city", "des_branch_code", "des_branch_name"],
        inplace=True
    )

    # Convert to numeric only once
    df = df.apply(pd.to_numeric, errors="coerce").fillna(0)

    # ==========================
    # Step 2: Aggregate
    # ==========================
    grouped = df.groupby(
        level=["org_branch_code", "service_type"]
    ).sum()

    # Prepare final records
    records = []
    for (branch, service_type), sub_df in grouped.groupby(level=["org_branch_code", "service_type"]):
        for type_ in ["Volume", "Billed Wt"]:
            if type_ in sub_df.columns.get_level_values("type"):
                # Filter by type and sum all destination branches
                type_mask = sub_df.columns.get_level_values("type") == type_
                total_sum = sub_df.loc[:, type_mask].sum().sum()
                records.append({
                    "org_branch_code": branch,
                    "service_type": service_type,
                    "type": type_,
                    "sum": round(total_sum, 3)
                })

    return pd.DataFrame(records)


# ==========================
# Example usage
# ==========================
org_branch = summarize_by_org_branch_type_service_optimized("data.csv")
org_branch.to_csv("org_summary.csv", index=False)

org_branch
```

```python
import pandas as pd

def summarize_by_des_branch_type_service(csv_path="data.csv"):
    df = pd.read_csv(
        csv_path,
        skiprows=1,
        header=[0, 1, 2, 3, 4, 5],
        index_col=[0, 1, 2, 3, 5, 6],
        low_memory=False
    )

    # Update index names
    df.index.set_names(
        ["org_zone", "org_region", "org_city", "org_branch_code", "service_type", "org_product"],
        inplace=True
    )
    df.columns.set_names(
        ["type", "des_zone", "des_region", "des_city", "des_branch_code", "des_branch_name"],
        inplace=True
    )

    # Ensure numeric
    df = df.apply(pd.to_numeric, errors="coerce").fillna(0)

    # ==========================
    # Optimized Destination Summary
    # ==========================
    records = []
    for type_ in ["Volume", "Billed Wt"]:
        if type_ in df.columns.get_level_values("type"):
            # Select only this type of columns
            type_df = df.loc[:, df.columns.get_level_values("type") == type_]

            # Sum across origins, grouped by destination branch
            summed = type_df.groupby(
                level=["des_branch_code"], axis=1
            ).sum()

            # Group by service_type across rows
            service_grouped = summed.groupby(df.index.get_level_values("service_type")).sum()

            # Build records
            for des_branch_code in summed.columns:
                for service_type in service_grouped.index.unique():
                    total_sum = summed.loc[df.index.get_level_values("service_type") == service_type, des_branch_code].sum()
                    records.append({
                        "des_branch_code": des_branch_code,
                        "service_type": service_type,
                        "type": type_,
                        "sum": round(total_sum, 3)
                    })

    return pd.DataFrame(records)


# ==========================
# Example usage
# ==========================
des_summary = summarize_by_des_branch_type_service("data.csv")
des_summary.to_csv("des_summary.csv", index=False)

des_summary
```

```python
import pandas as pd

# Load the office_location.csv file
office_df = pd.read_csv("office_location.csv")

office_df
```

```python
# Group by lat/lon and aggregate office codes as a list
lat_lon_offices = office_df.groupby(['lat', 'lon'])['office'].agg(list).reset_index()

# For each (lat, lon) shared by multiple offices, print the count and the codes
shared_lat_lon_count = 0
unique_lat_lon_count = 0
for _, row in lat_lon_offices.iterrows():
    if len(row['office']) > 1:
        print(f"({row['lat']}, {row['lon']}): count = {len(row['office'])}, offices = {row['office']}")
        shared_lat_lon_count += 1
    elif len(row['office']) == 1:
        unique_lat_lon_count += 1

print(f"Total (lat, lon) shared by multiple offices: {shared_lat_lon_count}")
print(f"Total (lat, lon) used by only one office: {unique_lat_lon_count}")
```
