```python
import pandas as pd
df = pd.read_csv(r'/Users/arup/Documents/sorter-clubbing-optimizer/origin_destination_flow.csv')

df
```

```python
# For the first 4 columns, fill down the last seen non-null value from row index 5 onwards
cols_to_fill = df.columns[:4]
for col in cols_to_fill:
    df.loc[5:, col] = df.loc[5:, col].ffill()

df
```

```python
# For the first 4 rows, fill right (horizontally) from the 6th column onwards with the last seen non-null value
import numpy as np

start_col = 5  # 0-based index, so 6th column
end_col = df.shape[1]
for row in range(4):
    last_val = np.nan
    for col in range(start_col, end_col):
        if pd.notnull(df.iat[row, col]):
            last_val = df.iat[row, col]
        else:
            df.iat[row, col] = last_val

df
```

```python
# From 6th row and 6th column (i.e., index 5 onwards), ensure all values are numbers, fix numbers written as text with commas, and fill missing values with 0

# Select the sub-DataFrame to process
numeric_block = df.iloc[5:, 5:]

# Remove commas and convert to numeric, coerce errors to NaN
numeric_block = numeric_block.applymap(lambda x: pd.to_numeric(str(x).replace(',', ''), errors='coerce'))

# Fill missing values with 0
numeric_block = numeric_block.fillna(0)

# Assign back to the original DataFrame
df.iloc[5:, 5:] = numeric_block

df
```

```python
df.to_csv('data.csv', index=False)
```

```python
import pandas as pd
from algorithms import filter_and_sum

def summarize_by_branch_type_mode(csv_path="data.csv"):
    df = pd.read_csv(csv_path, header=None, low_memory=False)
    branches = df.iloc[6:, 3].unique()

    types = ["Volume", "Billed Wt"]
    modes = ["Air", "Ground"]

    records = []
    for branch in branches:
        for mode in modes:
            for type_ in types:
                total = filter_and_sum(
                    type_=type_,
                    mode=mode,
                    org_branch=branch,
                    csv_path=csv_path
                )
                records.append({
                    "org_branch": branch,
                    "type": type_,
                    "mode": mode,
                    "sum": total
                })

    return pd.DataFrame(records)

# Run the summarizer
branch = summarize_by_branch_type_mode("data.csv")

# (Optional) Save to CSV if you want a file
branch.to_csv("branch_summary.csv", index=False)

# Check first few rows
branch
```
